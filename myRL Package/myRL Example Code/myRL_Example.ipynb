```python
### Import specific classes and functions from each package

from ddpg_package import DDPG
from myrl_package import myRL, TrainingPlan, TuningPlan, TestingPlan
```

```python
### Example use

# Create an object 'myRL_hub' from the myRL class
myRL_hub = myRL()


### TRAINING


# Create a training plan object from the TrainingPlan object
trainingPlan_v1 = TrainingPlan(algo=DDPG, eps=1500, its=2, name="training-plan DDPG v1", env_name="BipedalWalker-v3")

# Add the training plan to the myRL object
myRL_hub.addTrainingPlan(trainingPlan_v1)

# Show the training plans stored in the myRL object
myRL_hub.showTrainingPlans()

# Run the training plan
myRL_hub.runTrainingPlan("training-plan DDPG v1")

# Get the results for only iteration 1, and download them as an excel file
myRL_hub.getTrainingPlanResults(name="training-plan DDPG v1", it=1, download=True)

# Display the results of the training plan
myRL_hub.displayTrainingResults(trainingPlanName="training-plan DDPG v1", mean_its=False, it=1, legend=True, legend_loc=3, moving_average=20)


### TUNING


# Create a tuning plan object from the TuningPlan object
tuningPlan_v1 = TuningPlan(trainingPlan=trainingPlan_v1, param="actor_lr", vals=[0.0005, 0.0002], name="tuning-plan DDPG v1")

# Add the tuning plan to the myRL object
myRL_hub.addTuningPlan(tuningPlan_v1)

# Show the tuning plans stored in the myRL object
myRL_hub.showTuningPlans()

# Run the tuning plan
myRL_hub.runTuningPlan("tuning-plan DDPG v1")

# Get the results for the specific value tested (0.0005), and don't download the results as an excel file
myRL_hub.getTuningPlanResults(name="tuning-plan DDPG v1", val=0.0005, download=False)

# Display the results of the training plan
myRL_hub.displayTuningResults(tuningPlanName="tuning-plan DDPG v1", mean_its=False, val=0.0002, legend=True, legend_loc=3, moving_average=20)


### TESTING


# Get the agents to be tested
Agent_1 = myRL_hub.getAgent(trainingPlanName = "training-plan DDPG v1", it=1)
Agent_2 = myRL_hub.getAgent(tuningPlanName = "tuning-plan DDPG v1", it=1, val=0.0005)

# Create the testing plans
testingPlan_1 = TestingPlan(name="testing-plan 1", env_name="BipedalWalker-v3", eps=10, agent=Agent_1)
testingPlan_2 = TestingPlan(name="testing-plan 2", env_name="BipedalWalker-v3", eps=10, agent=Agent_2)

# Add the testing plans to the myRL object
myRL_hub.addTestingPlan(testingPlan=testingPlan_1)
myRL_hub.addTestingPlan(testingPlan=testingPlan_2)

# Show the testing plans stored in the myRL object
myRL_hub.showTestingPlans()

# Run the test
myRL_hub.runTestingPlan(name="testing-plan 1")

# Get and download the results of the test
myRL_hub.getTestingPlanResults(name="testing-plan 1", download=True)

# Display the results of the testing plan
myRL_hub.displayTestingResults(testingPlanName="testing-plan 1", legend=True, legend_loc=3)
```
